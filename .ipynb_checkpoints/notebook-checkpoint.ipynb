### Como é a distribuição dos base stats conforme os tipos?

```python
def plot_base_stat_distribution_by_type(stat: str, type_1_or_2: int):
    stat_mean = pkm.groupby('Type' + ' ' + str(type_1_or_2))[stat].mean()
    sns.barplot(x = stat_mean, y = stat_mean.index, palette = 'inferno', order = stat_mean.sort_values(ascending = False).index)
```

### Teste de Normalidade: Teste de Kolmogorov-Smirnov

Referências para os testes de normalidade:

http://www.mat.uc.pt/~cmtm/ECwww/TestesNP.pdf

https://www.statology.org/normality-test-python/

https://psicometriaonline.com.br/como-testar-a-normalidade-da-amostra-com-kolgomorov-smirnov-e-shapiro-wilk/

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kstest.html

Utilizaremos o teste de Kolmogorov-Smirnov ao invés do teste de Shapiro-Wilk devido ao grande tamanho da amostra.

```python
n = len(pkm['HP'])
print(f'Tamanho da amostra: {n}')
```

Teste de hipóteses dos testes de normalidade:

$H_{0}$: a distribuição é normal.

$H_{A}$: a distribuição não é normal.

```python
for stat in pkm_stats:
    normality_test = kstest(pkm[stat], norm.cdf)
    print(f'{stat}: {normality_test}')
    print(f'pvalue < 0.05: {normality_test.pvalue < 0.05}')
    print(f'Normality: {not normality_test.pvalue < 0.05}\n')
```

- Como todos os pvalue são menores 0.05, podemos rejeitar a hipótese nula do teste de Kolmogorov-Smirnov a uma significância de $\alpha = 5\%$ para todos os stats. Dessa forma, podemos afirmar que as distribuições de todos os stats não são normais. No entanto, pelo fato das amostras serem muito grandes, a hipótese nula tende a ser muito rejeitada, de tal forma que é melhor analisarmos outras formas de teste de normalidade, como os gráficos Q-Q.

### Gráficos Q-Q

Referência: https://towardsdatascience.com/understand-q-q-plot-using-simple-python-4f83d5b89f8f

Documentação: https://www.statsmodels.org/dev/generated/statsmodels.graphics.gofplots.qqplot.html

#### 1ª opção: utilizando os dados não normalizados e a opção line = 's'

```python
fig, axis = plt.subplots(2, 3, figsize = (15,15))

plot_hp = sm.qqplot(pkm['HP'], line='s', ax = axis[0,0])
plot_atk = sm.qqplot(pkm['Attack'], line='s', ax = axis[0,1])
plot_def = sm.qqplot(pkm['Defense'], line='s', ax = axis[0,2])
plot_sp_atk = sm.qqplot(pkm['Sp. Atk'], line='s', ax = axis[1,0])
plot_sp_def = sm.qqplot(pkm['Sp. Def'], line='s', ax = axis[1,1])
plot_speed = sm.qqplot(pkm['Speed'], line='s', ax = axis[1,2])

plt.show()
```

As distribuições, de fato, não são normais, pois não se aproximam totalmente da linha de 45º. As que mais se aproximam, no entanto, são as de 'Attack' e de 'Speed'.

#### 2ª opção: normalizando os dados

```python
pkm_stats
```

```python
pkm_norm = dict(zip(pkm_stats, [[None]]*len(pkm_stats)))
for stat in pkm_stats:
    stat_mean = pkm[stat].mean()
    stat_std = pkm[stat].std()
    pkm_norm[stat] = pkm[stat].apply(lambda x: (x - stat_mean)/stat_std)
pkm_norm = pd.DataFrame(pkm_norm)
```

```python
pkm_norm = pd.DataFrame(pkm_norm)
```

```python
pkm_norm
```

```python
fig, axis = plt.subplots(2, 3, figsize = (15,15))

plot_norm_hp = sm.qqplot(pkm_norm['HP'], line='45', ax = axis[0,0])
plot_norm_atk = sm.qqplot(pkm_norm['Attack'], line='45', ax = axis[0,1])
plot_norm_def = sm.qqplot(pkm_norm['Defense'], line='45', ax = axis[0,2])
plot_norm_sp_atk = sm.qqplot(pkm_norm['Sp. Atk'], line='45', ax = axis[1,0])
plot_norm_sp_def = sm.qqplot(pkm_norm['Sp. Def'], line='45', ax = axis[1,1])
plot_norm_speed = sm.qqplot(pkm_norm['Speed'], line='45', ax = axis[1,2])

plt.show()
```
